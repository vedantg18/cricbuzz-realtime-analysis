## Fetching Match Sceduled Series

```python
import requests
import pandas as pd

url = "https://cricbuzz-cricket.p.rapidapi.com/series/v1/7607"

headers = {
	"x-rapidapi-key": "73badbcd63mshf9da4a5b470b009p118844jsn7c808fe48912",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

print(response.json())
```

```python
import pandas as pd
from pandas import json_normalize

# Step 1: Load DataFrame
df = pd.DataFrame(response.json()['matchDetails'])

# Step 2: Clean column names
df.columns = df.columns.str.strip()

# Step 3: Initialize list to collect flattened rows
flattened_rows = []

# Step 4: Loop through each row
for index, row in df.iterrows():
    match_details_map = row['matchDetailsMap']
    
    # Check if match_details_map is a valid dict
    if isinstance(match_details_map, dict):
        key = match_details_map.get('key', None)
        match_list = match_details_map.get('match', [])
        
        if isinstance(match_list, list):
            for match in match_list:
                flattened_match = match.get('matchInfo', {})
                flattened_match['key'] = key
                flattened_rows.append(flattened_match)

# Step 5: Create flattened DataFrame
final_df = pd.DataFrame(flattened_rows)

# Step 6: Export to CSV
final_df.to_csv('scheduled.csv', index=False)
```
